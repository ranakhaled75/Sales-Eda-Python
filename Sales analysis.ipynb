```python
import pandas as pd
import numpy as np
```

```python
sales=pd.read_csv("sale.csv")
sales
```

```python
sales.info()
```

```python
sales["order_date"]=pd.to_datetime(sales["order_date"])
sales["ship_date"]=pd.to_datetime(sales["ship_date"])
```

```python
sales.isnull().sum()
```

```python
sales.describe()
```

```python
sales["Year"]=sales["order_date"].dt.year
sales["Month"]=sales["order_date"].dt.month
```

# 1. **Total Sales by Region**: #

```python
sales.groupby("region")[['sales']].sum()
```

# 2. **Profit Margin by Category**: #

```python
sales.groupby("category")[['profit_margin']].sum()
```

# 3.**Monthly Sales Over Time**: #

```python
sales.groupby("Month")[['sales']].sum()
```

# 4. **Distribution of Discounts**: #


```python
sales["discount"].describe()
```

# 5. **Sales by Category and Subcategory**:

```python
sales.groupby(["category",'subcategory'])[['sales']].sum()
```

# 6. **Profit vs. Sales by Category**:

```python
sales.groupby("category")[['sales','profit']].sum()
```

# 7. **Total Quantity Sold by State**:


```python
sales.groupby("state")[['quantity']].sum()
```

# 8. **Average Profit Margin by Region**:


```python
sales.groupby('region')[['profit']].mean()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(sales["region"],sales["sales"],color='red')
plt.xlabel("days")
plt.title("Sales over days")
plt.show()
```

# 10. **Top 10 Products by Sales**:


```python
sales.groupby('product_name')['sales'].sum().sort_values(ascending =False).head(10)
```
